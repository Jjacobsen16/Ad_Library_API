```python
# this makes the demographic dist column into python datatypes (a list of dicts) to be used for manipulation
#df.loc = abaility to access a group of rows and columns by label(s) or a boolean array. (components of a string in this case)
#dropna will remove those with no values(NAN)
#apply(literal_eval) converts the json to dicts



df.loc[:,'fixed_dems'] = df.loc[:,'demographic_distribution'].dropna().apply(literal_eval)
```

```python
#these functions dig out the percentages, add them up and return them as floats 
#this could probably be more elegantly combined into one function 

def male(x):
    male_percentage = 0.0
    for i in x:
        if i['gender'] == "male":
            male_percentage += float(i['percentage'])
    return male_percentage
            
def female(x):
    female_percentage = 0.0
    for i in x:
        if i['gender'] == "female":
            female_percentage += float(i['percentage'])
    return female_percentage    
```

```python
#assigning the new columns, via the function call

df.loc[:,'percentage_male'] = df['fixed_dems'].dropna().apply(male)
df.loc[:,'percentage_female'] = df['fixed_dems'].dropna().apply(female)
```

```python
#to make the dominant_gender column

df.loc[:,'dominant_gender'] = np.where(df['percentage_male'] < df['percentage_female'], 'female','male')


```

```python
#this is the same thing for age: first the function, then the call 
#this will output a cell that has the age groups (age_1 is 18-24, age_2 25-34 and so on, age_5 is 65+), formatted as dictionaries 

def age(x):
    age_ = 0.0
    age_1 = 0.0
    age_2 = 0.0
    age_3 = 0.0
    age_4 = 0.0
    age_5 = 0.0
    age_6 = 0.0
    for i in x:
        if i['age'] == "13-17":
            age_0 += float(i['percentage'])        
        elif i['age'] == "18-24":
            age_1 += float(i['percentage'])
        elif i['age'] == "25-34":
            age_2 += float(i['percentage'])
        elif i['age'] == "35-44":
            age_3 += float(i['percentage'])
        elif i['age'] == "45-54":
            age_4 += float(i['percentage'])      
        elif i['age'] == "55-64":
            age_5 += float(i['percentage']) 
        elif i['age'] == "65+":
            age_6 += float(i['percentage'])
    return dict(age_0 = age_0, age_1 = age_1, age_2 = age_2, age_3 = age_3, age_4 = age_4, age_5 = age_5, age_6 = age_6)
```

```python
def region(x):
    england = 0.0
    wales = 0.0
    scotland = 0.0
    northern_ireland = 0.0
    for i in x:
        if i['region'] == "England":
            england += float(i['percentage'])
        elif i['region'] == "Wales":
            wales += float(i['percentage'])
        elif i['region'] == "Scotland":
            scotland += float(i['percentage'])
        elif i['region'] == "Northern Ireland":
            northern_ireland += float(i['percentage'])            

    return dict(england = england, wales = wales, scotland = scotland, northern_ireland = northern_ireland)


```

```python
df.loc[:,'regions'] = df['fixed_region'].apply(region)
```
